```python
# Install library
!pip -q install music21

from music21 import stream, note, midi

# Create a simple melody
melody = stream.Stream()

notes = ['C4', 'D4', 'E4', 'G4', 'A4', 'G4', 'E4', 'D4', 'C4']

for n in notes:
    melody.append(note.Note(n, quarterLength=1))

# Save as MIDI file
midi_file = midi.translate.streamToMidiFile(melody)
midi_file.open('generated_music.mid', 'wb')
midi_file.write()
midi_file.close()

print("Music generated successfully!")
print("File saved as generated_music.mid")
```
